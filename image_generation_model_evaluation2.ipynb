```python
pipe = FluxPipeline.from_pretrained("black-forest-labs/FLUX.1-schnell", torch_dtype=torch.bfloat16)
#pipe.enable_model_cpu_offload()
pipe = pipe.to("cpu")
pipe.save_pretrained(pretrained_model_path+ "/FLUX.1-schnell/")
```

```python
pipe1 = StableDiffusionPipeline.from_pretrained(pretrained_model_path+ "/FLUX.1-schnell/")
```

```python
for i in range(50):
  prompt = dataset2["Prompt"][i]
  image = pipe(
    prompt,
    guidance_scale=0.0,
    num_inference_steps=4,
    max_sequence_length=256,
    generator=torch.Generator("cpu").manual_seed(0)
).images[0]
  image.save(output_path+"dataset2"+"/FLUX.1-schnell/"+f"{prompt}.png")
```
